```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
import matplotlib.cm as cm

graus_de_liberdade = [1, 3, 5, 11, 23]
x = np.linspace(0.01, max(graus_de_liberdade) * 2.5, 500)

plt.figure(figsize=(11, 7))
plt.title('Densidade da Distribuição ($\chi^2$)', fontsize=23)
plt.xlabel('Valor de $\chi^2$', fontsize=20)
plt.ylabel('Densidade', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.minorticks_on()

colormap = cm.get_cmap('ocean', len(graus_de_liberdade))

for i, grau in enumerate(graus_de_liberdade):
    cor = colormap(0.1 + (i / (len(graus_de_liberdade) - 1)) * 0.7)
    plt.plot(x, chi2.pdf(x, grau), label=f'$k = {grau}$', linewidth=2, color=cor)

plt.legend(title='Graus de Liberdade $k$', loc='upper right', frameon=True, shadow=True, fontsize=13, title_fontsize=14)
plt.ylim(0, 0.27)
plt.xlim(0, 30)

plt.tight_layout() # Ajusta o layout para evitar sobreposição
plt.savefig('qui2.pdf')
```
